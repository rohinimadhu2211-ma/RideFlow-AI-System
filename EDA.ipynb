```python
#VISUALIZATIONS
```

```python
#FARE PRICE DISTRIBUTION
```

```python
plt.figure(figsize=(8,5))

sns.histplot(

    df["fare_price"],

    bins=30,

    kde=True
)

plt.title(
    "Fare Price Distribution"
)

plt.xlabel(
    "Fare Price"
)

plt.show()
```

```python
#DRIVER RATING DISTRIBUTION
```

```python
plt.figure(figsize=(8,5))

sns.histplot(

    df["driver_rating"],

    bins=20,

    color="green"
)

plt.title(
    "Driver Rating Distribution"
)

plt.xlabel(
    "Driver Rating"
)

plt.show()
```

```python
#SURGE MULTIPLIER ANALYSIS
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x=df["surge_multiplier"]
)

plt.title(
    "Surge Multiplier Analysis"
)

plt.show()
```

```python
#ETA DISTRIBUTION
```

```python
plt.figure(figsize=(8,5))

sns.histplot(

    df["estimated_eta_min"],

    bins=30,

    color="red"
)

plt.title(
    "Estimated ETA Distribution"
)

plt.xlabel(
    "ETA Minutes"
)

plt.show()
```

```python
#WEATHER DISTRIBUTION
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x=df["weather"]
)

plt.title(
    "Weather Distribution"
)

plt.show()
```

```python
#PICKUP ZONE DISTRIBUTION
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    x=df["pickup_zone"]
)

plt.title(
    "Pickup Zone Distribution"
)

plt.show()
```

```python
#FARE PRICE VS SURGE
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(

    x=df["fare_price"],

    y=df["surge_multiplier"]
)

plt.title(
    "Fare Price vs Surge Multiplier"
)

plt.show()
```

```python
#CORRELATION HEATMAP
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(

    df.corr(numeric_only=True),

    cmap="coolwarm",

    annot=False
)

plt.title(
    "Correlation Heatmap"
)

plt.show()
```
